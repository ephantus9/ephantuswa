```python
# we will create a password checker to ensure the password has atleast 8 letters and a digit
print ("Hello! Welcome to password strength checker")

password=" "
valid= False 

while not valid:
    password = input("Enter password: ")
    
    # Check the password is at least 8 characters
    if len(password) < 8:
        print(" Password is too short. Must be at least 8 characters")
        continue
    
    # Check if the password contains at least one digit
    contains_digit = False
    for char in password:
        if char.isdigit():
            contains_digit = True
            break
            
    if not contains_digit:
            print("Password must contain at least one digit")
            continue
        
    valid = True 
    print(" Password accepted")
```

```python
print("Welcome to the Password Strength Checker")

valid = False

while not valid:
    password = input("Enter the password: ")

    # Check if password is at least 8 characters
    if len(password) < 8:
        print(" Password is too short. Must be at least 8 characters")
        continue

    # Check if password contains at least one digit
    contains_digit = False
    for char in password:
        if char.isdigit():
            contains_digit = True
            break

    if not contains_digit:
        print(" Password must contain at least one digit")
        continue

    valid = True
    print(" Password accepted")
```
