# Inheritance 
- A way to create a new class (child) that inherits attributes and methods from an existing class (parent).

```python
# Lets create a saving account -  limit the number of withdrawal to 3
class SavingsAccount(BankAccount):
    def __init__(self,owner,balance=0,withdrawal_limit=3):
        super(). __init__(owner,balance)    #This serve as a connection to the parent class
        self.withdrawal_limit = withdrawal_limit
        self.withdrawals_made = 0

    def withdraw(self,amount):
        if self.withdrawals_made >= self.withdrawal_limit:
            print("Transaction Error - withdraw limit reached")
        else:
            super().withdraw(amount) #super refers to the base class/we can also use it as BankAccount
            self.withdrawals_made += 1
```

### ABSTRACTION

> It hides complex implementation details and exposes only what is necessary

```python
from abc import ABC, abstractmethod
```

```python
class Account (ABC):
    def __init__(self,owner,balance=0):
        self.owner = owner
        self._balance = balance # Using protected attribute - convention

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self,amount):
        pass

    def get_balance(self):
        print(f"current balance: {self._balance:,.2f}")
```

```python
# create a BankAccount class from the abstract base class
class BankAccountAbs(Account):
    def deposit(self,amount):
        if amount > 0:
            self._balance += amount
            print(f"deposited: ${amount:,.2f}")
        else:
            print("Invalid amount attempted")

    def withdraw(self,amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew ${amount:,.2f}, Current Balance: ${self._balance:,.2f}")
        else:
            print(f"Invalid transaction or insufficient funds")
```

##  Activity 5:  
**Abstraction (Defining a Common Interface)**  
-Task: Create an abstract PersonBase class with an abstract introduce method.  
-Make Person and Student inherit from it, implementing introduce to describe themselves (e.g., name/age for Person, adding student)  

```python
from abc import ABC, abstractmethod

class PersonBase(ABC):
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    @abstractmethod
    def introduce(self):
        pass

class Person(PersonBase): 
    def introduce(self):
        return (f"Hello everyone, my name is {self.name} and I am {self.age} years old.")

class Student(Person):
    def __init__(self, name: str, age: int, student_id: str):
        super().__init__(name, age)
        self.student_id = student_id

    def introduce(self):
        base = super().introduce().rstrip(".")
        return (f"{base}, and my student ID is {self.student_id}.")
```

```python
person = Person("Sonyika", 45)
print(person.introduce())
```

```python
student = Student("Asha", 20, "S12345")
```

```python
print(student.introduce())
```

***Generator functions***

used for more complicated generators  
It uses the `yield` statement rather than the `return` statement.  

```python
G1 = [n ** 2 for n in range(12)]
```

```python
def gen():
    for n in range (12):
        yield n**2
    
```

```python
gen2=gen()
```

```python
#Another example
def divide_by(divide_by, numbers):
    for number in range(numbers+1):
        yield round(number/divide_by, 2) # we have rounded off to 2decimal places
    
```

```python
div = divide_by(3, 10)
```

```python
div
```

```python
for num in div:
    print(num, end=', ')
```

```python
# wwe sill `try` and except block to handle failure gracefully
def convert_float(x):
    try:
        return float (x)
    except:
        return x
```

```python
def convert_float(x):
    try:
        return float(x)

    except ValueError:
        return x
```

```python
convert_float('something')
```

```python
convert_float((1,2))
```

```python
# we can suppress multiple errors using tuples
def convert_float(x):
    try:
        return float(x)

    except (TypeError, ValueError):
        return x
```

```python
convert_float((1,2))
```

```python
#We use the `finally` statement to get the code executed whether the `try` block or not
try:
    number=int(input("Enter number:"))
    result = 10/number
    print('Result:', result)
except ZeroDivisionError:
    print('You cannot divide by zero')
except ValueError:
    print('Please enter a valid number')
finally:
    print('Thank you for using this code')
    
```

```python
#We can use else statement to run code if tge try does not fail
try:
    age = int(input('Enter your age:'))
except ValueError:
    print('This is not a valid age')
else:
    print(f'You entered:{age}')
finally:
    print('Thanks for using the age checker')
```
